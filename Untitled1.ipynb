```python
import pandas as pd
import matplotlib.pyplot as plt

arquivo = "t 0,1 us.txt"

df = pd.read_csv(arquivo, delim_whitespace=True)

wavelength = df["Wavelength"]

# média dos shots
intensity = df.drop(columns=["Wavelength"]).mean(axis=1)

plt.figure(figsize=(10,5))
plt.plot(wavelength, intensity)

plt.xlabel("Comprimento de onda (nm)")
plt.ylabel("Intensidade")
plt.title("Espectro LIBS médio")

plt.grid(True)
plt.show()
```

Começo do projeto

```python
import os
import pandas as pd
import glob

# 1. Extrair o arquivo .rar
# O comando !unrar extrai o conteúdo do arquivo P03.rar
!unrar x P03.rar -o+

# 2. Localizar todos os arquivos .esf extraídos
# Procuramos em todas as subpastas criadas pela extração
caminho_arquivos = glob.glob('**/*.esf', recursive=True)
print(f"Total de arquivos encontrados: {len(caminho_arquivos)}")

# Listas para armazenar os dados
dados_terceira_coluna = []
primeira_coluna = None

for arquivo in caminho_arquivos:
    try:
        # 3. Ler o arquivo pulando as 40 primeiras linhas
        # Usamos sep=None e engine='python' para detectar automaticamente o separador (espaço ou tab)
        # O parâmetro usecols=[0, 2] seleciona a 1ª e a 3ª coluna
        df_temp = pd.read_csv(arquivo, skiprows=40, header=None, sep=None, engine='python', usecols=[0, 2])

        # Guardamos a primeira coluna apenas uma vez (do primeiro arquivo lido)
        if primeira_coluna is None:
            primeira_coluna = df_temp[0]

        # Adicionamos a terceira coluna à nossa lista para média posterior
        dados_terceira_coluna.append(df_temp[2])

    except Exception as e:
        print(f"Erro ao processar {arquivo}: {e}")

# 4. Criar o DataFrame final
# Transformamos a lista de colunas em um DataFrame onde cada coluna é um arquivo
df_colunas_3 = pd.concat(dados_terceira_coluna, axis=1)

# Calculamos a média entre as linhas (axis=1)
media_coluna_3 = df_colunas_3.mean(axis=1)

# Criamos o DataFrame de resultado com a primeira coluna original e a média
df_final = pd.DataFrame({
    'Primeira_Coluna': primeira_coluna,
    'Media_Terceira_Coluna': media_coluna_3
})

# 5. Visualizar e Salvar
print("\nProcessamento concluído. Primeiras linhas do resultado:")
print(df_final.head())

# Salvar em um novo arquivo CSV se desejar
df_final.to_csv('resultado_processado.csv', index=False)
print("\nArquivo 'resultado_processado.csv' salvo com sucesso.")
```

```python
import pandas as pd
import plotly.express as px

# 1. Carregar o arquivo processado
df = pd.read_csv('resultado_processado.csv')

# 2. Criar o gráfico interativo com Plotly
# Configuramos os labels diretamente no comando
fig = px.line(
    df,
    x='Primeira_Coluna',
    y='Media_Terceira_Coluna',
    labels={
        'Primeira_Coluna': 'Comprimento da Onda',
        'Media_Terceira_Coluna': 'Intensidade'
    },
    title='Espectro de Intensidade - Média P03'
)

# 3. Customizar a aparência para o estilo "Clean/Científico"
fig.update_traces(line=dict(color='bluek', width=1)) # Linha preta e fina
fig.update_layout(
    template='plotly_white', # Fundo branco
    xaxis_title='Comprimento da Onda',
    yaxis_title='Intensidade',
    hovermode='x' # Mostra o valor exato ao passar o mouse no eixo X
)

# 4. Exibir o gráfico
# No Colab, isso abrirá uma janela onde você pode usar o mouse para:
# - Clicar e arrastar para dar zoom
# - Duplo clique para voltar ao normal
# - Usar a barra de ferramentas no canto superior direito para salvar como PNG
fig.show()
```
